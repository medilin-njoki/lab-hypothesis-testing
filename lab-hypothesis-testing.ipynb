```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
len(df[df['Type 1'] == 'Dragon']['HP'])
```

```python
# H0: Dragon mu HP >= Grass mu HP
# H1: Dragon mu HP < Grass mu HP
# We consider a sample of 40 Pokemons from each type
# Significance level = 0.05

dragon = df[df['Type 1'] == 'Dragon']['HP'].sample(30)
grass = df[df['Type 1'] == 'Grass']['HP'].sample(30)

st.ttest_ind(dragon, grass, alternative='less')
```

We are not able to reject H0 hypothesis: no significant evidence

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# H0: Legendary mu stats = Non Legendary mu stats
# H1: Legendary mu stats != Non Legendary mu stats
# We consider a sample of 40 Pokemons from each type
# Significant level = 0.05

non_legendary = df.loc[df['Legendary'] == False, ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sample(40)

legendary = df.loc[df['Legendary'] == True, ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sample(40)

st.ttest_ind(non_legendary, legendary, alternative='two-sided')
```

We reject the null hypothesis because all the p-values are lower than the significance level

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
# Define the hospital coordinates
hospital_lon, hospital_lat = -122.0, 37.0

# Define the school coordinates
school_lon, school_lat = -118.0, 34.0

# Compute Euclidean distance to hospital and assign to new column
df['distance_to_hospital'] = np.sqrt(
    (df.longitude - hospital_lon)**2 +
    (df.latitude  - hospital_lat)**2
)

# Compute Euclidean distance to school and assign to new column
df['distance_to_school'] = np.sqrt(
    (df.longitude - school_lon)**2 +
    (df.latitude  - school_lat)**2
)

# Create the boolean column
df["close_to_services"] = (
    (df["distance_to_hospital"] < 0.5) |
    (df["distance_to_school"]  < 0.5)
)

df
```

```python
# H0: houses close to either a school or a hospital price >  houses far from either a school or a hospital
# H1: houses close to either a school or a hospital price <= houses far from either a school or a hospital
# Significant level = 0.05

far = df[df['close_to_services'] == False]['median_house_value'].sample(40)
close = df[df['close_to_services'] == True]['median_house_value'].sample(40)

st.ttest_ind(close, far, alternative='less')
```

we fail to reject the null hypothesis because the p-value is greater than the significant level
